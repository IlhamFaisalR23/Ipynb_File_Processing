# Instruksi 3 : Definisi ulang read_file agar dapat membatasi baris yang dicetak

```python
def read_file(csv_file_name, column_start=0, column_end=None, limit=None):
  with open(csv_path + csv_file_name, mode='r') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0

        header = next(csv_reader)
        print(f'Column names: ')
        print(', '.join(header[column_start:column_end]))

        for row in csv_reader:
            if limit is not None and line_count >= limit:
                break

            print(', '.join(row[column_start:column_end]))
            line_count += 1

        print(f'Processed {line_count} lines (excluding header).')
  csv_file.close()
```

```python
read_file('employees.csv',3,5,10)
```

# Instruksi 4 : Definisi ulang read_file agar dapat melakukan sortir pada kolom pertama

```python
def sort_key(item):
    return item[0]

def read_file(csv_file_name, column_start, column_end, limit, sort=0):
    with open(csv_path + csv_file_name, mode='r') as csv_file:
        reader = csv.reader(csv_file)

        header = next(reader)
        print(f'{header[column_start]}, {header[column_end]}')

        data = [row[column_start:column_end + 1] for row in reader][:limit]

        if sort == 1:
            data.sort(key=sort_key)
        elif sort == -1:
            data.sort(key=sort_key, reverse=True)

        for row in data:
            print(f'{row[0]}, {row[-1]}')

        print(f'Processed {len(data) + 1} lines include header.')

    csv_file.close()
```

```python
read_file('employees.csv', 3, 5, 10, -1)
```

```python
read_file('employees.csv', 3,5,10,1)
```

# Instruksi 5 : Definisi baru find_file menampilkan data berdasarkan kata kunci

```python
def find_file(csv_file_name, index_column, keyword):
    with open(csv_path + csv_file_name, mode='r') as csv_file:
        reader = csv.reader(csv_file)
        header = next(reader)
        print(', '.join(header), {header[index_column]})

        for row in reader:
            if keyword in row[index_column]:
                print(', '.join(row), {row[index_column]})

    csv_file.close()
```

# Instruksi 8 : Definisi procedure insert_file (memasukkan data baru ke posisi baris paling akhir)

```python
def insert_file(csv_file_name, list_data):
    with open(csv_path + csv_file_name, mode='a', newline='') as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow(list_data)
    csv_file.close()
```

# Instruksi 9 : Definisi update_file (mengubah satu data yang sudah ada berdasarkan kolom pertama)

```python
def update_file(csv_file_name, keyword, index_column, new_value):
    # Membaca data dari file CSV
    rows = []

    with open(csv_path + csv_file_name, 'r') as csv_file:
        reader = csv.reader(csv_file)
        headers = next(reader)  # Membaca header jika ada
        for row in reader:
            rows.append(row)

    # Memperbarui nilai pada kolom index_column yang berbeda
    for row in rows:
        if row[0] == keyword:  # Memastikan pencarian berdasarkan kolom pertama
            row[index_column] = str(new_value)  # Mengubah nilai pada kolom yang diinginkan

    # Menulis kembali data ke dalam file CSV
    with open(csv_path + csv_file_name, 'w', newline='') as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow(headers)  # Menulis kembali header
        writer.writerows(rows)  # Menulis kembali baris yang sudah diperbarui
```

```python
update_file('countries.csv', 'ID', 2, 3)
```

```python
read_file('countries.csv')
```

# Instruksi 10 : Definisi delete_file (menghapus baris yang akan dihapus berdasarkan kolom paling pertama)

```python
def delete_file(csv_file_name, keyword):
  with open(csv_path + csv_file_name, mode='r') as csv_file:
    reader = csv.reader(csv_file)
    data = [row for row in reader if row[0] != keyword]
  csv_file.close()

  with open(csv_path + csv_file_name, mode='w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerows(data)
  csv_file.close()
```

```python
delete_file('countries.csv', 'ZL')
```
